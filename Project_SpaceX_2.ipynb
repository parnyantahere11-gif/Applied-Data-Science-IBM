```python

## Read the Data

import pandas as pd

df = pd.read_csv('/Users/taherehehsan/preprocessed_dataset.csv')
```

```python

df.head()
```

```python


# This preprocessed dataset is ready for modeling, as it has no Non_Null values, also all values are numerical.
# But we prefer to double check it.

```

```python

## So we have a safe dataset to be a good input for our ML model.
```

```python

## Define X & y

X = df.drop('Class', axis = 1)

y = df['Class']
```

```python


#######  Building a Logistic Regression model

# As this is a supervised Learning model we should apply train_set, test_set on dataset.

## Train_set, Test_set in Sikit_learn Library

!pip3 install scikit-learn

from sklearn.model_selection import train_test_split
```

```python

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.20, 
                                                    random_state = 101)
```

```python

## Train The Model

from sklearn.linear_model import LogisticRegression
```

```python

logmodel = LogisticRegression()

logmodel.fit(X_train, y_train)
```

```python

## Predict the Model 

predictions = logmodel.predict(X_test)

predictions
```

```python

y_test
```

```python

## Evaluation: To what extend it predict correctly?
# (TP, FP, TN, FN)

from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, predictions)
```

```python

## If we define class = 1 as a successful (positive) delivery 
## and class = 0 as a failed (negative) delivery, 
## then when the model predicts correctly, we call it True; otherwise, we call it False.

## In this case, 
# the model correctly predicts successful delivery 7 times — True Positives (TP)
# It never predicts a successful delivery incorrectly, so False Negatives (FN) are 0.
# On the other hand,
# the model correctly predicts failed delivery 10 times — True Negatives (TN)
# it incorrectly predicts one failed delivery as successful, False Positive (FP).

# The higher the number of True Positives and True Negatives, 
# the better the model’s performance.

# The model’s performance is acceptable, 
# but more evaluation is needed to reach a higher accuracy, ideally close to 100%.

```

```python

## Accuracy Score

from sklearn.metrics import accuracy_score

accuracy_score(y_test, predictions, normalize = False)  # number of True predicted
```

```python

accuracy_score(y_test, predictions, normalize = True)  # Percent of True predicted
```

```python

## Classification Report

from sklearn.metrics import classification_report

print(classification_report(y_test, predictions)) 
```

```python

## Evaluations

from sklearn.metrics import confusion_matrix , accuracy_score , classification_report
```

```python

confusion_matrix(y_test, predictions)
```

```python

accuracy_score(y_test, predictions, normalize = False)
```

```python

accuracy_score(y_test, predictions, normalize = True)
```

```python
	
print(classification_report(y_test,predictions))
```

```python

## We can improve the predictions to some extend by using GridSearch and changing hyperparameters.
```

```python

## Grid Search for Support Vector Machine Parameters
# Hyperparameter

svm_1 = SVC()

# we change C=1.0 and kernel='rbf'
```

```python

# Allows us to test parameters of classification algorithms and find the best one

from sklearn.model_selection import GridSearchCV
```

```python

svm_1 = SVC()
```

```python

parameters = {'C': [0.5, 1, 1.5], 'kernel': ['linear', 'rbf', 'sigmoid']}
```

```python

## We can change 2 other parameters: 
# cv defines how many folds the training data is split into for cross-validation.
# verbose means “verbosity level” — how detailed the output messages are during training.

svm_cv = GridSearchCV(svm_1, parameters, cv = 4, verbose = 3)
```

```python

svm_cv.fit(X_train, y_train)
```

```python

## Grid Search Best Hyperparameters

print('tuned hyperparameters : (best hyperparameters)', svm_cv.best_params_)
```

```python

## Tune Best Hyperparameters

svm_1 = SVC( C = 1, kernel = 'linear')
```

```python

svm_1.fit(X_train, y_train)
```

```python

predictions_1 = svm_1.predict(X_test)
```

```python

confusion_matrix(y_test, predictions_1)
```

```python

accuracy_score(y_test, predictions_1, normalize = False)
```

```python

accuracy_score(y_test, predictions_1, normalize = True)
```

```python

print(classification_report(y_test, predictions_1))
```
